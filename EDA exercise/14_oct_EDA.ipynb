```
#Exloring Matplot Libb


x= np.linspace(0,10,20)
fig,axes = plt.subplots(1,2,figsize=(12,5))
axes[0].plot(x,x**2,x,np.exp(x))
axes[0].set_title('Normal Scale')
axes[0].grid(True)

axes[1].plot(x,x**2,x,np.exp(x))
axes[1].set_title('Log Scale')
axes[1].set_yscale('Log')
axes[1].grid(True)
plt.show()
```

```
#Twin Axes


fix,ax1 = plt.subplots()

ax1.plot(x,x**2,color='b')
ax1.set_ylabel('Area')

ax2  = ax1.twinx()
ax2.plot(x,x**3,color='r')
ax1.set_ylabel('Volume')
```

```
n = np.array([0,1,2,3,4,5])
x= np.linspace(0,10,20)
fig,axes = plt.subplots(1,4,figsize=(10,5))


axes[0].scatter(x,x**2+np.random.randn(len(x)))
axes[0].set_label('scatter')

axes[1].step(n,n**2,color='r')
axes[1].set_label('step')

axes[2].bar(n,n**2)
axes[2].set_label('bar')

axes[3].fill_between(x,x**2,x**3)
axes[3].set_label('fill')
plt.show()
```
