```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


from sklearn.datasets import load_iris 
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names) 
label = pd.DataFrame(list(map(lambda x : iris.target_names[x], iris.target)), columns=['Species']) 
iris = pd.concat([data, label], axis=1) 
iris.head()
```

```
fig,axs = plt.subplots(2,2,figsize=(12,8))
sns.distplot(iris['sepal length (cm)'],ax=axs[0][0],axlabel='Sepal Length in cms')
sns.distplot(iris['sepal width (cm)'],ax=axs[0][1],axlabel='Sepal Width in cms')
sns.distplot(iris['petal length (cm)'],ax=axs[1][0],axlabel='Petal Length in cms')
sns.distplot(iris['petal width (cm)'],ax=axs[1][1],axlabel='Petal width in cms')
```

```
iris.info()
```

```
fig,axs = plt.subplots(2,2,figsize=(12,8))
sns.boxplot(iris['sepal length (cm)'],ax=axs[0][0])
sns.boxplot(iris['sepal width (cm)'],ax=axs[0][1])
sns.boxplot(iris['petal length (cm)'],ax=axs[1][0])
sns.boxplot(iris['petal width (cm)'],ax=axs[1][1])
```
